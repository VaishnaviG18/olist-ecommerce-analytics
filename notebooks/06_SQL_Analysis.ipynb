# 6. HAVING Clause

The `HAVING` clause is used to filter grouped data after applying the `GROUP BY` clause. While the `WHERE` clause filters individual rows before grouping, the `HAVING` clause filters the summarized groups based on aggregate conditions.

It is commonly used in business reporting to identify high-performing states, top-selling product categories, and frequently used payment methods.

```python
query = """
SELECT
    customer_state,
    COUNT(DISTINCT order_id) AS total_orders
FROM ecommerce
GROUP BY customer_state
HAVING COUNT(DISTINCT order_id) > 1000
ORDER BY total_orders DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    product_category_name,
    ROUND(SUM(price + freight_value),2) AS total_revenue
FROM ecommerce
GROUP BY product_category_name
HAVING SUM(price + freight_value) > 100000
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    payment_type,
    ROUND(AVG(order_value),2) AS average_order_value
FROM ecommerce
GROUP BY payment_type
HAVING AVG(order_value) > 500
ORDER BY average_order_value DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    product_category_name,
    ROUND(AVG(review_score),2) AS average_review_score
FROM ecommerce
GROUP BY product_category_name
HAVING AVG(review_score) > 4.5
ORDER BY average_review_score DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    customer_state,
    ROUND(AVG(delivery_days),2) AS average_delivery_days
FROM ecommerce
GROUP BY customer_state
HAVING AVG(delivery_days) < 10
ORDER BY average_delivery_days;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    seller_id,
    COUNT(product_id) AS products_sold
FROM ecommerce
GROUP BY seller_id
HAVING COUNT(product_id) > 100
ORDER BY products_sold DESC;
"""

pd.read_sql(query, conn)
```

# 7. CASE WHEN

The `CASE WHEN` statement is used to apply conditional logic in SQL. It evaluates one or more conditions and returns a value based on the first matching condition.

It is commonly used to:
- Categorize customers
- Classify orders
- Create business labels
- Generate custom reporting fields

```python
query = """
SELECT
    order_id,
    order_value,
    CASE
        WHEN order_value < 100 THEN 'Low Value'
        WHEN order_value BETWEEN 100 AND 500 THEN 'Medium Value'
        ELSE 'High Value'
    END AS order_category
FROM ecommerce
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    order_id,
    delivery_days,
    late_delivery,
    CASE
        WHEN late_delivery = 1 THEN 'Late Delivery'
        ELSE 'On Time'
    END AS delivery_status
FROM ecommerce
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    customer_unique_id,
    customer_order_count,
    CASE
        WHEN customer_order_count = 1 THEN 'New Customer'
        ELSE 'Repeat Customer'
    END AS customer_type
FROM ecommerce
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    payment_installments,
    CASE
        WHEN payment_installments = 1 THEN 'Single Payment'
        WHEN payment_installments BETWEEN 2 AND 3 THEN 'Short Installment'
        WHEN payment_installments BETWEEN 4 AND 6 THEN 'Medium Installment'
        ELSE 'Long Installment'
    END AS installment_category
FROM ecommerce
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    review_score,
    CASE
        WHEN review_score >= 4 THEN 'Positive'
        WHEN review_score = 3 THEN 'Neutral'
        ELSE 'Negative'
    END AS review_sentiment
FROM ecommerce
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    product_category_name,
    price,
    CASE
        WHEN price < 50 THEN 'Budget'
        WHEN price BETWEEN 50 AND 200 THEN 'Standard'
        ELSE 'Premium'
    END AS product_segment
FROM ecommerce
LIMIT 10;
"""

pd.read_sql(query, conn)
```

# 8. Subqueries

A subquery is a query nested inside another SQL query. It allows us to use the result of one query as the input for another query.

Subqueries are commonly used to compare values against averages, identify top-performing customers, filter records dynamically, and solve complex business problems.

```python
query = """
SELECT
    order_id,
    order_value
FROM ecommerce
WHERE order_value >
(
    SELECT AVG(order_value)
    FROM ecommerce
);
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    product_id,
    product_category_name,
    price
FROM ecommerce
WHERE price >
(
    SELECT AVG(price)
    FROM ecommerce
)
LIMIT 20;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT DISTINCT
    customer_unique_id,
    customer_total_spending
FROM ecommerce
WHERE customer_total_spending >
(
    SELECT AVG(customer_total_spending)
    FROM ecommerce
)
ORDER BY customer_total_spending DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    order_id,
    delivery_days
FROM ecommerce
WHERE delivery_days =
(
    SELECT MAX(delivery_days)
    FROM ecommerce
);
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    customer_state,
    ROUND(SUM(price + freight_value),2) AS total_revenue
FROM ecommerce
GROUP BY customer_state
HAVING SUM(price + freight_value) >
(
    SELECT AVG(state_revenue)
    FROM
    (
        SELECT
            SUM(price + freight_value) AS state_revenue
        FROM ecommerce
        GROUP BY customer_state
    )
)
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    customer_unique_id,
    COUNT(DISTINCT order_id) AS total_orders
FROM ecommerce
GROUP BY customer_unique_id
HAVING COUNT(DISTINCT order_id) >
(
    SELECT AVG(order_count)
    FROM
    (
        SELECT
            COUNT(DISTINCT order_id) AS order_count
        FROM ecommerce
        GROUP BY customer_unique_id
    )
)
ORDER BY total_orders DESC;
"""

pd.read_sql(query, conn)
```

# 9. Common Table Expressions (CTEs)

A Common Table Expression (CTE) is a temporary named result set created using the `WITH` clause. It simplifies complex SQL queries by breaking them into smaller, more readable parts.

CTEs improve query readability, maintainability, and debugging. They are widely used in business reporting, data analysis, and ETL processes.

```python
query = """
WITH state_revenue AS
(
    SELECT
        customer_state,
        ROUND(SUM(price + freight_value),2) AS total_revenue
    FROM ecommerce
    GROUP BY customer_state
)

SELECT *
FROM state_revenue
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
WITH customer_spending AS
(
    SELECT
        customer_unique_id,
        MAX(customer_total_spending) AS total_spending
    FROM ecommerce
    GROUP BY customer_unique_id
)

SELECT *
FROM customer_spending
ORDER BY total_spending DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
WITH delivery_summary AS
(
    SELECT
        customer_state,
        ROUND(AVG(delivery_days),2) AS avg_delivery_days
    FROM ecommerce
    GROUP BY customer_state
)

SELECT *
FROM delivery_summary
ORDER BY avg_delivery_days ASC;
"""

pd.read_sql(query, conn)
```

```python
query = """
WITH category_revenue AS
(
    SELECT
        product_category_name,
        ROUND(SUM(price + freight_value),2) AS total_revenue
    FROM ecommerce
    GROUP BY product_category_name
)

SELECT *
FROM category_revenue
ORDER BY total_revenue DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
WITH state_revenue AS
(
    SELECT
        customer_state,
        SUM(price + freight_value) AS total_revenue
    FROM ecommerce
    GROUP BY customer_state
)

SELECT
    customer_state,
    ROUND(total_revenue,2) AS total_revenue
FROM state_revenue
WHERE total_revenue >
(
    SELECT AVG(total_revenue)
    FROM state_revenue
)
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
WITH payment_summary AS
(
    SELECT
        payment_type,
        ROUND(AVG(order_value),2) AS avg_order_value
    FROM ecommerce
    GROUP BY payment_type
)

SELECT *
FROM payment_summary
ORDER BY avg_order_value DESC;
"""

pd.read_sql(query, conn)
```

# 10. Window Functions

Window functions perform calculations across a set of rows related to the current row without collapsing the results into a single row.

Unlike `GROUP BY`, window functions return a value for every row while still allowing aggregate calculations.

Window functions are widely used for:
- Ranking
- Running totals
- Moving averages
- Previous and next row analysis
- Top-N analysis

```python
query = """
SELECT
    order_id,
    order_value,
    ROW_NUMBER() OVER(
        ORDER BY order_value DESC
    ) AS row_number
FROM ecommerce
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    order_id,
    order_value,
    RANK() OVER(
        ORDER BY order_value DESC
    ) AS order_rank
FROM ecommerce
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    order_id,
    order_value,
    DENSE_RANK() OVER(
        ORDER BY order_value DESC
    ) AS dense_rank
FROM ecommerce
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    customer_state,
    order_id,
    order_value,
    RANK() OVER(
        PARTITION BY customer_state
        ORDER BY order_value DESC
    ) AS state_rank
FROM ecommerce;
"""

pd.read_sql(query, conn).head(20)
```

```python
query = """
SELECT
    order_purchase_timestamp,
    price,
    SUM(price) OVER(
        ORDER BY order_purchase_timestamp
    ) AS running_revenue
FROM ecommerce
LIMIT 20;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    order_purchase_timestamp,
    order_value,
    LAG(order_value) OVER(
        ORDER BY order_purchase_timestamp
    ) AS previous_order_value
FROM ecommerce
LIMIT 20;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    order_purchase_timestamp,
    order_value,
    LEAD(order_value) OVER(
        ORDER BY order_purchase_timestamp
    ) AS next_order_value
FROM ecommerce
LIMIT 20;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    customer_state,
    order_id,
    order_value,
    ROUND(
        AVG(order_value) OVER(
            PARTITION BY customer_state
        ),
        2
    ) AS state_average_order_value
FROM ecommerce;
"""

pd.read_sql(query, conn).head(20)
```
