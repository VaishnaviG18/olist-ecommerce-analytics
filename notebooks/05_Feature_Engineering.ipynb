```python
'''
## 1. Convert Date Columns

Before creating time-based features, all date-related columns must be converted to the datetime format. This enables date calculations such as delivery duration, purchase month, purchase year, and order approval time.
'''
```

```python
# ==========================================================
# Convert Date Columns
# ==========================================================

date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'shipping_limit_date',
    'review_creation_date',
    'review_answer_timestamp'
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col])

print("Date columns converted successfully!")
```

```python
'''
## 2. Date & Time Features

Time-based features help analyze customer purchasing behavior across different months, weekdays, quarters, and hours. These features are valuable for identifying seasonal trends, peak shopping periods, and customer activity patterns.
'''
```

```python
# ==========================================================
# Date & Time Features
# ==========================================================

df['purchase_year'] = df['order_purchase_timestamp'].dt.year

df['purchase_month'] = df['order_purchase_timestamp'].dt.month

df['purchase_day'] = df['order_purchase_timestamp'].dt.day

df['purchase_day_name'] = df['order_purchase_timestamp'].dt.day_name()

df['purchase_hour'] = df['order_purchase_timestamp'].dt.hour

df['purchase_quarter'] = df['order_purchase_timestamp'].dt.quarter

df['is_weekend'] = df['purchase_day_name'].isin(['Saturday', 'Sunday'])
```

```python
df[['purchase_year',
    'purchase_month',
    'purchase_day',
    'purchase_day_name',
    'purchase_hour',
    'purchase_quarter',
    'is_weekend']].head()
```

```python
'''
## 3. Delivery Features

Delivery-related features provide insights into the efficiency of the order fulfillment process. By calculating delivery durations and comparing actual delivery dates with estimated delivery dates, we can identify delays, measure logistics performance, and evaluate customer experience.

These engineered features will support future business analysis, SQL queries, and predictive machine learning models.
'''
```

```python
# ==========================================================
# Delivery Days
# ==========================================================

df['delivery_days'] = (
    df['order_delivered_customer_date']
    - df['order_purchase_timestamp']
).dt.days
```

```python
# ==========================================================
# Approval Days
# ==========================================================

df['approval_days'] = (
    df['order_approved_at']
    - df['order_purchase_timestamp']
).dt.days
```

```python
# ==========================================================
# Estimated Delivery Gap
# ==========================================================

df['estimated_delivery_gap'] = (
    df['order_estimated_delivery_date']
    - df['order_delivered_customer_date']
).dt.days
```

```python
# ==========================================================
# Late Delivery Flag
# ==========================================================

df['late_delivery'] = (
    df['order_delivered_customer_date']
    > df['order_estimated_delivery_date']
)
```

```python
'''
## 4. Revenue & Order Features

Revenue-based features help measure the financial value of each order and categorize orders based on their monetary contribution. These engineered features are useful for identifying high-value customers, analyzing purchasing behavior, and supporting future predictive models.

The features created in this section will help answer business questions such as:
- Which orders generate the highest revenue?
- How many high-value orders are placed?
- What is the distribution of order values?
'''
```

```python
# ==========================================================
# Total Order Value
# ==========================================================

df['total_order_value'] = df['price'] + df['freight_value']
```

```python
# ==========================================================
# High Value Order
# ==========================================================

average_order_value = df['total_order_value'].mean()

df['high_value_order'] = (
    df['total_order_value'] > average_order_value
)
```

```python
# ==========================================================
# Revenue Bucket
# ==========================================================

df['revenue_bucket'] = pd.qcut(
    df['total_order_value'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
# ==========================================================
# Actual Order Value
# ==========================================================

df['order_value'] = df.groupby('order_id')['total_order_value'].transform('sum')
```

```python
average_order_value = df['order_value'].mean()

df['high_value_order'] = df['order_value'] > average_order_value
```

```python
df['revenue_bucket'] = pd.qcut(
    df['order_value'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
'''
## 5. Customer Features

Customer-based feature engineering helps understand customer purchasing behavior and loyalty. These features provide insights into how frequently customers place orders, how much they spend over time, and whether they are repeat buyers.

These features are valuable for customer segmentation, retention analysis, SQL reporting, and predictive machine learning models.
'''
```

```python
# ==========================================================
# Customer Order Count
# ==========================================================

df['customer_order_count'] = (
    df.groupby('customer_unique_id')['order_id']
      .transform('nunique')
)
```

```python
# ==========================================================
# Repeat Customer Flag
# ==========================================================

df['repeat_customer'] = (
    df['customer_order_count'] > 1
)
```

```python
# ==========================================================
# Customer Total Spending
# ==========================================================

df['customer_total_spending'] = (
    df.groupby('customer_unique_id')['order_value']
      .transform('sum')
)
```

```python
# ==========================================================
# Average Order Value Per Customer
# ==========================================================

df['average_order_value_per_customer'] = (
    df['customer_total_spending']
    / df['customer_order_count']
)
```

```python
'''
## 6. Payment Features

Payment-related feature engineering helps categorize customer payment behavior. Instead of using raw installment numbers, grouping them into meaningful categories makes analysis easier and provides useful features for business reporting and predictive models.

These features can help answer questions such as:
- Do customers prefer single payments or installments?
- Which installment groups contribute more revenue?
- Does installment behavior affect customer spending?
'''
```

```python
# ==========================================================
# Installment Category
# ==========================================================

df['installment_category'] = pd.cut(
    df['payment_installments'],
    bins=[0, 1, 3, 6, 24],
    labels=[
        'Single Payment',
        'Short Installment',
        'Medium Installment',
        'Long Installment'
    ]
)
```
