# Business Question 1: Sales Trend Analysis

## Objective

Analyze how sales have changed over time to identify growth patterns, seasonal trends, and business performance.

### Why is this important?

- Helps understand whether sales are increasing or decreasing.
- Identifies peak sales periods.
- Supports inventory planning and marketing campaigns.
- Assists management in forecasting future sales.

```python
orders_master["order_purchase_timestamp"] = pd.to_datetime(
    orders_master["order_purchase_timestamp"]
)
```

```python
orders_master["Year_Month"] = (
    orders_master["order_purchase_timestamp"]
    .dt.to_period("M")
)
```

```python
monthly_orders = (
    orders_master
    .groupby("Year_Month")["order_id"]
    .nunique()
    .reset_index()
)

monthly_orders.columns = [
    "Year_Month",
    "Total_Orders"
]

monthly_orders
```

```python
monthly_orders["Year_Month"] = (
    monthly_orders["Year_Month"]
    .astype(str)
)
```

```python
monthly_orders_filtered = monthly_orders[
    monthly_orders["Year_Month"] < "2018-09"
]
```

```python
plt.figure(figsize=(14,6))

plt.grid(True, alpha=0.3)

plt.plot(
    monthly_orders_filtered["Year_Month"],
    monthly_orders_filtered["Total_Orders"],
    marker="o",
    linewidth=2
)

plt.title("Monthly Order Trend (Excluding Incomplete Months)")
plt.xlabel("Month")
plt.ylabel("Number of Orders")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
monthly_orders.tail()
```

## Business Insights

### Key Findings

- The number of orders increased steadily throughout 2017, indicating strong business growth.
- The highest order volume was observed in **November 2017**, likely influenced by seasonal events such as Black Friday.
- Order volume remained consistently high during most of 2018, suggesting stable customer demand.
- The months of **September 2018** and **October 2018** were excluded from the visualization because they contain incomplete data and would otherwise create a misleading downward trend.

### Business Recommendation

- Prepare inventory and marketing campaigns for high-demand periods such as November.
- Investigate the strategies used during peak months and replicate them in future campaigns.
- When performing time-series analysis, always verify whether the latest periods contain complete data before drawing business conclusions.

# Business Question 2: Monthly Revenue Analysis

## Objective

Analyze how the company's monthly revenue has changed over time to identify growth trends, seasonal patterns, and overall business performance.

### Why is this important?

- Measures the company's financial growth.
- Identifies high-revenue months.
- Helps in budgeting and forecasting.
- Supports marketing and inventory planning.

```python
print("Total Rows:", len(orders_master))
print("Unique Orders:", orders_master["order_id"].nunique())
```

```python
orders_master[
    ["order_id", "payment_value", "order_item_id", "payment_sequential"]
].head(10)
```

```python
monthly_revenue = (
    revenue_data
    .groupby("Year_Month")["payment_value"]
    .sum()
    .reset_index()
)

monthly_revenue.columns = [
    "Year_Month",
    "Total_Revenue"
]
```

```python
monthly_revenue = monthly_revenue[
    monthly_revenue["Year_Month"] < "2018-09"
]
```

```python
monthly_revenue.tail()
```

## Business Insights

### Key Findings

- Revenue showed steady growth throughout 2017, indicating increasing customer demand and business expansion.
- The highest monthly revenue was recorded in **November 2017**, which aligns with the peak in order volume. This suggests that seasonal shopping events (such as Black Friday) had a significant positive impact.
- Revenue remained consistently above **1 million** for most of 2018, indicating stable financial performance.
- Although revenue fluctuated slightly from month to month, there was no major decline after the business reached maturity.

### Business Recommendations

- Increase inventory and marketing efforts before high-revenue months such as November.
- Analyze the promotional strategies used during peak months and replicate successful campaigns.
- Maintain customer engagement throughout the year to sustain stable monthly revenue.
- Continue monitoring monthly revenue trends to identify seasonal opportunities and respond to changes in customer demand.

# Business Question 3: Customer Satisfaction Analysis

## Objective

Analyze customer review scores to understand overall customer satisfaction and identify opportunities for improving the shopping experience.

### Why is this important?

- Measures customer satisfaction.
- Identifies whether customers are generally happy with their purchases.
- Helps improve product quality, delivery service, and customer support.

```python
print("=" * 90)
print("REVIEW SCORE DISTRIBUTION")
print("=" * 90)

review_distribution = (
    orders_master["review_score"]
    .value_counts()
    .sort_index()
)

print(review_distribution)
```

```python
plt.figure(figsize=(8,5))

ax = review_distribution.plot(kind="bar")

plt.title("Distribution of Customer Review Scores")
plt.xlabel("Review Score")
plt.ylabel("Number of Reviews")

for i, value in enumerate(review_distribution):
    plt.text(i, value + 500, f"{value:,}", ha="center", fontsize=10)

plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

## Business Insights

### Key Findings

- The majority of customers gave a **5-star rating (66,343 reviews)**, indicating a high level of customer satisfaction.
- **4-star ratings (22,319)** are the second most common, suggesting that most customers had a positive shopping experience.
- **1-star ratings (15,428)** are noticeably higher than both 2-star and 3-star ratings. This indicates that while many customers were highly satisfied, a significant number were also very dissatisfied.
- **2-star (4,162)** and **3-star (9,894)** ratings are relatively less common, showing that customers generally had either a very positive or a very negative experience rather than a neutral one.

### Business Recommendations

- Investigate the reasons behind the high number of 1-star reviews by analyzing delivery delays, damaged products, incorrect orders, and customer complaints.
- Identify the products and sellers receiving the highest number of low ratings to improve quality control.
- Continue the practices that contribute to high customer satisfaction and 5-star reviews.
- Use customer feedback to improve delivery services and overall shopping experience.

# Business Question 7: Delivery Performance Analysis

## Objective

Analyze the delivery performance by calculating the time taken to deliver orders to customers.

### Why is this important?

- Measures logistics efficiency.
- Identifies delivery delays.
- Improves customer satisfaction.
- Helps optimize supply chain operations.
- Supports better delivery planning.

```python
orders_master["order_purchase_timestamp"] = pd.to_datetime(
    orders_master["order_purchase_timestamp"]
)

orders_master["order_delivered_customer_date"] = pd.to_datetime(
    orders_master["order_delivered_customer_date"]
)
```

```python
orders_master["delivery_days"] = (
    orders_master["order_delivered_customer_date"]
    -
    orders_master["order_purchase_timestamp"]
).dt.days
```

```python
delivery_data = orders_master[
    orders_master["delivery_days"].notna()
]
```

```python
print("="*90)
print("DELIVERY TIME STATISTICS")
print("="*90)

delivery_data["delivery_days"].describe()
```

```python
average_delivery = delivery_data["delivery_days"].mean()

print(f"Average Delivery Time: {average_delivery:.2f} Days")
```

```python
plt.figure(figsize=(10,6))

plt.hist(
    delivery_data["delivery_days"],
    bins=30,
    edgecolor="black"
)

plt.title("Distribution of Delivery Time")
plt.xlabel("Delivery Days")
plt.ylabel("Number of Orders")

plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,2))

plt.boxplot(
    delivery_data["delivery_days"],
    vert=False
)

plt.title("Delivery Time Box Plot")
plt.xlabel("Delivery Days")

plt.tight_layout()
plt.show()
```

## Business Insights

- The average delivery time is approximately **12 days**, indicating efficient logistics for most orders.
- The majority of deliveries are completed within **6 to 15 days**.
- The delivery time distribution is positively skewed, meaning a small number of orders experience significant delays.
- Several outliers with delivery times exceeding 100 days were identified, suggesting exceptional operational or logistical issues.
- Improving these delayed deliveries can enhance customer satisfaction and overall service quality.

# Business Question 8: Order Status Analysis

## Objective

Analyze the distribution of order statuses to evaluate overall operational performance.

### Why is this important?

- Understand the percentage of completed and cancelled orders.
- Measure order fulfillment performance.
- Identify operational bottlenecks.
- Improve customer satisfaction.

```python
order_status_data = (
    orders_master[
        ["order_id", "order_status"]
    ]
    .drop_duplicates()
)
```

```python
order_status_count = (
    order_status_data["order_status"]
    .value_counts()
)

print("=" * 90)
print("ORDER STATUS DISTRIBUTION")
print("=" * 90)

print(order_status_count)
```

```python
plt.figure(figsize=(10,6))

ax = order_status_count.plot(
    kind="bar",
    color="steelblue",
    edgecolor="black"
)

plt.title("Order Status Distribution")
plt.xlabel("Order Status")
plt.ylabel("Number of Orders")

plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.3)

# Add value labels
for i, value in enumerate(order_status_count):
    ax.text(
        i,
        value,
        f"{value:,}",
        ha="center",
        va="bottom",
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

```python
order_status_percentage = (
    order_status_count /
    order_status_count.sum() * 100
).round(2)

print("=" * 90)
print("ORDER STATUS PERCENTAGE")
print("=" * 90)

print(order_status_percentage)
```

## Business Insights

- Approximately **97% of all orders were successfully delivered**, indicating excellent order fulfillment performance.
- Less than **1% of orders were canceled**, suggesting a low cancellation rate.
- A very small percentage of orders were marked as unavailable, highlighting effective inventory management.
- Only a few orders remained in processing, invoiced, or shipped status, indicating efficient logistics and operational workflows.
- Overall, the company demonstrates a highly reliable order fulfillment process with minimal operational disruptions.

# Business Question 9: Top Customers Analysis

## Objective

Identify the customers who generated the highest revenue for the business.

### Why is this important?

- Identify high-value customers.
- Understand customer purchasing behavior.

```python
customer_revenue = (
    orders_master[
        [
            "customer_unique_id",
            "order_id",
            "payment_sequential",
            "payment_value"
        ]
    ]
    .drop_duplicates()
)
```
